#### Leemos el csv obtenido de thinkspeak, lo volvemos tipo dataframe y los mostramos

```python
import pandas as pd
datos=pd.read_csv('data_thinkspeak.csv')
df = pd.DataFrame(datos)
df.head()
#len(df)
#df
```

```python
df['field1']  #nos obtiene lacomunda llamada 'field1'
df.loc[0,'field1']   #obtenemos el valor de la fila '0' y la  comunda 'field1'
l=df['field1']>59  #se aplica la condicional para cada valor de la columna
l
df.loc[df['field1']>79,'field1'] #hacemos un filtro a la columna df.loc[boleano_o_int, 'name_columna']
#l,type(l)
```

```python
alto = df.loc[0,'field1']
df.loc[0,'dfield1']=0
alto,type(alto)
```

```python
#accedemos a la columna
alto = df.loc[0,'field1']
df.loc[0,'dfield1']=0

for i in range(1,len(df)):
    if df.loc[i,'field1']>df.loc[i-1,'field1'] and df.loc[i,'field1']>alto:
        df.loc[i,'dfield1']=df.loc[i,'field1']-alto
        alto=df.loc[i,'field1']
    else:
        df.loc[i,'dfield1']=0

df
```

```python
#exportamos a un csv
df.reset_index().to_csv('final.csv',header=True, index=False)
```
